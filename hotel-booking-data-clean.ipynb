```python
import pandas as pd
import numpy as np

# read in all our data
hotel_data = pd.read_csv("/kaggle/input/hotel-booking-demand/hotel_bookings.csv")

# set seed for reproducibility
np.random.seed(0) 
```

```python
hotel_data.head()
```

```python
missing_values_count = hotel_data.isnull().sum()

missing_values_count[0:10]
```

```python
total_cells = np.product(hotel_data.shape)
total_missing = missing_values_count.sum()
percent_missing = (total_missing/total_cells) * 100
print(percent_missing)
```

```python
hotel_data.dropna()
```

```python
col_na_drop = hotel_data.dropna(axis=1)
col_na_drop.head()
```

```python
print("Columns in original dataset: %d \n" % hotel_data.shape[1])
print("Columns with na's dropped: %d" % col_na_drop.shape[1])
```
